## **Temel Fonksiyonlar**

AÅŸaÄŸÄ±daki kod bloÄŸunda tanÄ±mladÄ±ÄŸÄ±mÄ±z `get_stats` ve `merge` fonksiyonlarÄ±, **`bpe-temeller.ipynb`** dosyasÄ±nda detaylÄ±ca incelediÄŸimiz ve algoritmanÄ±n Ã§ekirdeÄŸini oluÅŸturan fonksiyonlarÄ±n **birebir aynÄ±sÄ±dÄ±r.**

MantÄ±k deÄŸiÅŸmemiÅŸtir, sadece bu fonksiyonlarÄ± ÅŸimdi daha bÃ¼yÃ¼k bir veri seti Ã¼zerinde kullanacaÄŸÄ±z:

* **`get_stats(ids)`:** Veri setini tarayarak yan yana gelen byte ikililerinin frekansÄ±nÄ± hesaplar (**Ä°statistikÃ§imiz**).
* **`merge(ids, pair, idx)`:** Belirlenen en sÄ±k Ã§ifti listede bulur ve yeni atanan token ID'si ile deÄŸiÅŸtirerek sÄ±kÄ±ÅŸtÄ±rmayÄ± yapar (**BirleÅŸtiricimiz**).

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

def merge(ids, pair, idx):
  newids = []
  i = 0
  while i < len(ids):
    if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
      newids.append(idx)
      i += 2
    else:
      newids.append(ids[i])
      i += 1
  return newids
```

##**Hiperparametre TanÄ±mlarÄ± ve Veri HazÄ±rlÄ±ÄŸÄ±**

Bu bÃ¶lÃ¼m, BPE algoritmasÄ±nÄ±n eÄŸitim sÄ±nÄ±rlarÄ±nÄ± Ã§izdiÄŸimiz ve veriyi iÅŸleme hazÄ±rladÄ±ÄŸÄ±mÄ±z konfigÃ¼rasyon aÅŸamasÄ±dÄ±r.

    * Standart UTF-8 kodlamasÄ±nda **0-255** arasÄ± deÄŸerler temel byte'larÄ± (karakter parÃ§alarÄ±nÄ±) temsil eder.
    * BelirlediÄŸimiz **276** deÄŸeri, algoritmanÄ±n bu temel setin Ã¼zerine **20 adet yeni "birleÅŸtirilmiÅŸ token"** ekleyerek duracaÄŸÄ±nÄ± belirtir. Bu deÄŸer artÄ±rÄ±ldÄ±kÃ§a metin daha fazla sÄ±kÄ±ÅŸÄ±r, ancak kelime daÄŸarcÄ±ÄŸÄ± (vocabulary) bÃ¼yÃ¼r.

2.  **`num_merges` (Ä°terasyon SayÄ±sÄ±):**
    * EÄŸitim dÃ¶ngÃ¼sÃ¼nÃ¼n kaÃ§ adÄ±m Ã§alÄ±ÅŸacaÄŸÄ±nÄ± belirleyen sayaÃ§tÄ±r.
    * FormÃ¼l: $\text{Hedef Boyut} - \text{Temel Boyut} = 276 - 256 = 20 iterasyon.


```python
vocab_size = 276
num_merges = vocab_size - 256
ids = list(dataset.encode("utf-8"))

merges = {}
for i in range(num_merges):
  stats = get_stats(ids)
  pair = max(stats, key=stats.get)
  idx = 256 + i

  print(f"{pair} Ã§ifti birleÅŸtirilerek yeni token {idx} oluÅŸturuluyor")

  ids = merge(ids, pair, idx)
  merges[pair] = idx

print(f"\nText UzunluÄŸu: {len(dataset)}, Token UzunluÄŸu: {len(list(dataset.encode('utf-8')))}")
print(f"SÄ±kÄ±ÅŸtÄ±rma OranÄ±: {len(list(dataset.encode('utf-8'))) / len(ids):.2f}X")
```
